# organizando a iteração

```python
nomes = ['01','05','08','12','15','19','22','26','29']
Radares = [5323,5226,5358,5357,5355,5356,5151,5205,5204,5313,5316,5315,5314,5476,5475,5474,5479,5478,5477,5480,5196,5195,5194,5197,5203,5202,5432,5354,5414,5413,5433,5098,5496,5129,5471,5470,5199,5198,5201,5200,5508,5507,5308,5307]
path = r'csv_radares-20200426T211130Z-001\\'

#para cada um dos arquivos
for name in nomes:
    #leio
    df = pd.read_csv(path + name + ".csv")
    
    #selecino os radares que desejo
    df_selec = pd.DataFrame()
    for i in range(len(Radares)):
        dftemp = df.loc[df["Local"]==Radares[i]]
        df_selec = pd.concat([df_selec,dftemp])
    # Temos o df com redares selecionados
    
    # Transformo as colunas em datetime
    # mudo o formato
    df_selec['Ano']=df_selec['Ano'].astype(str)
    df_selec['Mes']=df_selec['Mes'].astype(str)
    df_selec['Dia']=df_selec['Dia'].astype(str)
    df_selec['Hora']=df_selec['Hora'].astype(str)
    df_selec['Minuto']=df_selec['Minuto'].astype(str)
    df_selec['Segundo']=df_selec['Segundo'].astype(str)
    #concatena o conteudo das colunas
    df_selec['Data']= df_selec['Dia'] + '/' + df_selec['Mes'] + '/' + df_selec['Ano'] + ' ' + df_selec['Hora'] + ':' + df_selec['Minuto'] + ':' + df_selec['Segundo'] 
    #passa pra datetime
    df_selec.Data = pd.to_datetime(df_selec.Data)
    #dropa as velhas colunas de data e hora
    df_selec = df_selec.drop(columns = ['Dia', 'Mes',"Ano","Hora", 'Minuto','Segundo'])
    
    
    #mudando a ordem das colunas para data fica na frente
    cols = df_selec.columns.tolist()
    dat = cols[-1]
    cols.remove('Data')
    cols.insert(1, dat)
    df_selec = df_selec[cols]
    df_selec.head()

    #reset o index
    df_selec.reset_index(drop=True,inplace=True)

    #cria o novo arquivo
    df_selec.to_csv('csv_radares-20200426T211130Z-001_selecionados\\' + name +'_selecionado.csv')
```
